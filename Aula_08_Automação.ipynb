Verificador de Temperatura:

```python
def verificar_temperatura():
    # Simula a leitura da temperatura do sistema
    # Em um ambiente real, você usaria bibliotecas específicas
    # para ler a temperatura do hardware (ex: psutil, wmi no Windows).
    try:
        temperatura_atual = float(input("Digite a temperatura atual em °C: "))
    except ValueError:
        print("Entrada inválida. Por favor, digite um número para a temperatura.")
        return

    print(f"Temperatura atual do sistema: {temperatura_atual:.2f}°C")

    if temperatura_atual > 30:
        print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C.")
    else:
        print("Temperatura do sistema normal.")

# Chama a função para verificar a temperatura
verificar_temperatura()
```

Organizador de Arquivos:

```python
import os
import shutil

def organizar_arquivos_por_extensao(diretorio_origem, diretorio_destino_base):
    """
    Organiza arquivos em um diretório baseando-se em suas extensões.

    Args:
        diretorio_origem (str): O caminho para o diretório onde os arquivos estão bagunçados.
        diretorio_destino_base (str): O caminho para o diretório onde as pastas organizadas serão criadas.
    """

    # 1. Tenta criar o diretório base para os arquivos organizados.
    # `exist_ok=True` previne um erro caso o diretório já exista.
    try:
        os.makedirs(diretorio_destino_base, exist_ok=True)
        print(f"Diretório de destino '{diretorio_destino_base}' garantido.")
    except OSError as e:
        # Captura e imprime erros relacionados à criação do diretório (ex: permissões).
        print(f"Erro ao criar o diretório de destino '{diretorio_destino_base}': {e}")
        return # Sai da função se não puder criar o diretório de destino.

    # 2. Lista todos os itens (arquivos e diretórios) no diretório de origem.
    for nome_arquivo in os.listdir(diretorio_origem):
        # 3. Constrói o caminho completo para o item atual no diretório de origem.
        caminho_completo_origem = os.path.join(diretorio_origem, nome_arquivo)

        # 4. Verifica se o item é um arquivo (e não um diretório).
        if os.path.isfile(caminho_completo_origem):
            # 5. Separa o nome do arquivo da sua extensão.
            # Ex: 'documento.pdf' -> ('documento', '.pdf')
            _, extensao = os.path.splitext(nome_arquivo)
            extensao = extensao.lower() # Converte a extensão para minúsculas para padronização.

            # 6. Determina o nome da pasta com base na extensão do arquivo.
            if extensao:
                # Remove o '.' inicial da extensão e adiciona um 's' (ex: '.pdf' -> 'pdfs').
                nome_pasta_extensao = extensao[1:] + 's'
            else:
                # Se o arquivo não tiver extensão, ele será movido para uma pasta 'sem_extensao'.
                nome_pasta_extensao = 'sem_extensao'

            # 7. Constrói o caminho completo para a pasta de destino específica da extensão.
            diretorio_destino_extensao = os.path.join(diretorio_destino_base, nome_pasta_extensao)

            # 8. Tenta criar a pasta específica para a extensão, se ela ainda não existir.
            try:
                os.makedirs(diretorio_destino_extensao, exist_ok=True)
            except OSError as e:
                # Captura e imprime erros na criação da subpasta.
                print(f"Erro ao criar a pasta '{diretorio_destino_extensao}': {e}")
                continue # Pula para o próximo arquivo se a subpasta não puder ser criada.

            # 9. Constrói o caminho completo para onde o arquivo será movido.
            caminho_completo_destino = os.path.join(diretorio_destino_extensao, nome_arquivo)

            # 10. Tenta mover o arquivo do diretório de origem para o diretório de destino adequado.
            try:
                shutil.move(caminho_completo_origem, caminho_completo_destino)
                print(f"Movido: '{nome_arquivo}' para '{diretorio_destino_extensao}'")
            except shutil.Error as e:
                # Erros específicos do shutil (ex: arquivo já existe no destino).
                print(f"Erro ao mover '{nome_arquivo}': {e}")
            except OSError as e:
                # Erros do sistema operacional (ex: permissão negada ao mover o arquivo).
                print(f"Erro de permissão ou outro erro do sistema ao mover '{nome_arquivo}': {e}")

    print("Organização de arquivos concluída!")

# --- Exemplo de Uso --- (Mantido para demonstração)
# 11. Define o diretório atual como base.
DIRETORIO_ATUAL = './'

# 12. Define os caminhos para os diretórios de teste e organizado.
DIRETORIO_TESTE = os.path.join(DIRETORIO_ATUAL, 'arquivos_baguncados')
DIRETORIO_ORGANIZADO = os.path.join(DIRETORIO_ATUAL, 'arquivos_organizados')

# 13. Garante que o diretório de teste exista.
os.makedirs(DIRETORIO_TESTE, exist_ok=True)

# 14. Cria alguns arquivos de teste apenas se eles não existirem, para evitar sobrescrever.
if not os.path.exists(os.path.join(DIRETORIO_TESTE, 'documento1.pdf')):
    with open(os.path.join(DIRETORIO_TESTE, 'documento1.pdf'), 'w') as f: f.write('pdf content')
    with open(os.path.join(DIRETORIO_TESTE, 'imagem1.jpg'), 'w') as f: f.write('jpg content')
    with open(os.path.join(DIRETORIO_TESTE, 'relatorio.docx'), 'w') as f: f.write('docx content')
    with open(os.path.join(DIRETORIO_TESTE, 'planilha.xlsx'), 'w') as f: f.write('xlsx content')
    with open(os.path.join(DIRETORIO_TESTE, 'foto.png'), 'w') as f: f.write('png content')
    with open(os.path.join(DIRETORIO_TESTE, 'arquivo_sem_extensao'), 'w') as f: f.write('no extension')
    print("Arquivos de teste criados em 'arquivos_baguncados'.")

# 15. Chama a função principal para organizar os arquivos.
# ATENÇÃO: Altere 'DIRETORIO_TESTE' e 'DIRETORIO_ORGANIZADO' para os seus caminhos reais se estiver usando em outro contexto.
organizar_arquivos_por_extensao(DIRETORIO_TESTE, DIRETORIO_ORGANIZADO)
```

Verificador de CEP:

```python
import requests
import json

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço a partir de um CEP.

    Args:
        cep (str): O número do CEP a ser consultado.

    Returns:
        dict or None: Um dicionário com os dados do endereço se a consulta for bem-sucedida,
                      ou None em caso de erro ou CEP inválido.
    """
    # 1. Remove caracteres não numéricos do CEP para garantir que apenas dígitos sejam processados.
    cep = ''.join(filter(str.isdigit, cep))

    # 2. Verifica se o CEP possui 8 dígitos após a limpeza.
    # Se não tiver 8 dígitos, é considerado inválido.
    if len(cep) != 8:
        print("Erro: CEP inválido. O CEP deve conter 8 dígitos numéricos.")
        return None

    # 3. Constrói a URL da API ViaCEP com o CEP fornecido.
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # 4. Faz a requisição HTTP GET para a API ViaCEP.
        # Um timeout de 5 segundos é adicionado para evitar que a requisição trave indefinidamente.
        response = requests.get(url, timeout=5)
        # 5. Levanta uma exceção para códigos de status HTTP 4xx (erros do cliente) ou 5xx (erros do servidor).
        response.raise_for_status()

        # 6. Converte a resposta JSON da API em um dicionário Python.
        data = response.json()

        # 7. Verifica se a API retornou um campo 'erro' com valor 'true'.
        # Isso indica que o CEP não foi encontrado ou é inválido do ponto de vista da API.
        if 'erro' in data and data['erro'] is True:
            print(f"Erro: CEP '{cep}' não encontrado ou inválido.")
            return None

        # 8. Retorna os dados do endereço se a consulta for bem-sucedida e sem erros.
        return data

    # 9. Bloco de tratamento de exceções para diversos tipos de erros que podem ocorrer durante a requisição:
    except requests.exceptions.HTTPError as http_err:
        # Erro de resposta HTTP (ex: 404 Not Found, 500 Internal Server Error).
        print(f"Erro HTTP ao consultar o CEP '{cep}': {http_err}")
    except requests.exceptions.ConnectionError as conn_err:
        # Erro de conexão de rede (ex: sem internet, DNS não resolvido).
        print(f"Erro de Conexão ao consultar o CEP '{cep}': {conn_err}\nVerifique sua conexão com a internet ou o status da API ViaCEP.")
    except requests.exceptions.Timeout as timeout_err:
        # A requisição excedeu o tempo limite definido.
        print(f"Tempo limite excedido ao consultar o CEP '{cep}': {timeout_err}\nA API ViaCEP demorou muito para responder.")
    except requests.exceptions.RequestException as req_err:
        # Um erro geral de requisição ocorreu (superclasse de outros erros requests).
        print(f"Ocorreu um erro inesperado na requisição para o CEP '{cep}': {req_err}")
    except json.JSONDecodeError as json_err:
        # A resposta da API não pôde ser decodificada como JSON válido.
        print(f"Erro ao decodificar a resposta JSON da API para o CEP '{cep}': {json_err}\nA resposta não é um JSON válido.")
    except Exception as e:
        # Captura qualquer outra exceção inesperada.
        print(f"Ocorreu um erro inesperado: {e}")

    # Em caso de qualquer erro, retorna None.
    return None

# --- Exemplo de Uso Interativo --- (Permite ao usuário digitar CEPs continuamente)
print("\n--- Consulta de CEP Interativa ---")
print("Digite 'sair' a qualquer momento para encerrar.")

while True:
    # 1. Solicita ao usuário que digite um CEP para consultar.
    cep_input = input("Digite o CEP para consultar: ")

    # 2. Verifica se o usuário digitou 'sair' (case-insensitive) para encerrar o loop.
    if cep_input.lower() == 'sair':
        print("Encerrando a consulta de CEP.")
        break

    # 3. Chama a função consultar_cep com o input do usuário.
    endereco = consultar_cep(cep_input)

    # 4. Se um endereço foi retornado (ou seja, não houve erro fatal e o CEP foi encontrado).
    if endereco:
        print("Endereço encontrado:")
        # 5. Verifica se a API retornou um erro (mesmo que o formato do CEP fosse válido).
        # A mensagem de erro já é impressa pela função consultar_cep neste caso, então apenas passa.
        if 'erro' in endereco and endereco['erro'] is True:
            pass
        else:
            # 6. Imprime o dicionário de endereço formatado como JSON com indentação para clareza.
            # ensure_ascii=False permite a impressão correta de caracteres acentuados.
            print(json.dumps(endereco, indent=4, ensure_ascii=False))

    # 7. Imprime uma linha separadora para melhorar a legibilidade entre as consultas.
    print("-" * 30)
```

Script de Alerta Autônomo:

```python
import time
import random

def verificar_condicao_critica(valor_atual, limite_critico):
    """
    Verifica se o valor atual excede um limite crítico.

    Args:
        valor_atual (float): O valor a ser verificado (ex: temperatura).
        limite_critico (float): O limite a partir do qual a condição é crítica.

    Returns:
        bool: True se a condição for crítica, False caso contrário.
    """
    print(f"[Monitor] Valor atual: {valor_atual:.2f} (Limite crítico: {limite_critico:.2f})")
    return valor_atual > limite_critico

def disparar_alerta(mensagem_alerta, tipo_alerta='console', email_destino=None):
    """
    Dispara um alerta, seja no console ou simulando um e-mail.

    Args:
        mensagem_alerta (str): A mensagem a ser incluída no alerta.
        tipo_alerta (str): 'console' para imprimir no console, 'email_simulado' para simular e-mail.
        email_destino (str, optional): Endereço de e-mail para o alerta simulado.
    """
    if tipo_alerta == 'console':
        print(f"\n{'='*30}\nALERTA CRÍTICO: {mensagem_alerta}\n{'='*30}\n")
    elif tipo_alerta == 'email_simulado':
        if email_destino:
            print(f"\n{'='*30}\nSIMULANDO ENVIO DE E-MAIL PARA {email_destino}:\nAssunto: ALERTA DE SISTEMA CRÍTICO\nCorpo: {mensagem_alerta}\n{'='*30}\n")
        else:
            print("Erro: E-mail de destino não fornecido para alerta simulado.")
    else:
        print(f"Tipo de alerta '{tipo_alerta}' não reconhecido. Mensagem: {mensagem_alerta}")

# --- Configurações do Sistema e Loop Autônomo ---

LIMITE_CRITICO = 30.0 # Exemplo: Temperatura máxima em °C
INTERVALO_CHECK_SEGUNDOS = 5 # Intervalo entre as verificações (em segundos)
EMAIL_ALERTA = "seu.email@exemplo.com" # Substitua pelo seu e-mail para testar a simulação

print(f"Iniciando monitoramento autônomo. Limite crítico: {LIMITE_CRITICO}°C.")
print(f"Verificação a cada {INTERVALO_CHECK_SEGUNDOS} segundos. Pressione Ctrl+C para parar.")

try:
    # Loop principal que executa o monitoramento continuamente
    while True:
        # Simula a leitura de um sensor (ex: temperatura)
        # Em um ambiente real, você leria dados de um sensor, API, etc.
        valor_do_sensor = random.uniform(20.0, 35.0) # Gera um valor aleatório entre 20 e 35

        if verificar_condicao_critica(valor_do_sensor, LIMITE_CRITICO):
            # Condição crítica atingida, dispara o alerta
            mensagem = f"Superaquecimento detectado! Temperatura: {valor_do_sensor:.2f}°C, acima do limite de {LIMITE_CRITICO}°C."

            # Dispara alerta no console
            disparar_alerta(mensagem, tipo_alerta='console')

            # Dispara alerta por e-mail simulado
            disparar_alerta(mensagem, tipo_alerta='email_simulado', email_destino=EMAIL_ALERTA)

        else:
            print("[Monitor] Condição normal.")

        # Aguarda o próximo ciclo de verificação
        time.sleep(INTERVALO_CHECK_SEGUNDOS)

except KeyboardInterrupt:
    print("\nMonitoramento encerrado pelo usuário.")
except Exception as e:
    print(f"Ocorreu um erro inesperado no loop principal: {e}")
```

### Script de Alerta Autônomo - Versão 2 (Refatorado)

Esta versão refatorada do script de alerta autônomo encapsula a lógica de monitoramento e alerta em uma classe `SistemaMonitoramento`. Isso torna o código mais modular, legível e fácil de estender.

**Principais mudanças:**
- **Classe `SistemaMonitoramento`**: Gerencia as configurações, o estado do monitoramento e as ações (verificação e disparo de alertas).
- **Métodos claros**: `_verificar_condicao_critica` e `_disparar_alerta` são métodos da classe.
- **Método `iniciar_monitoramento`**: Contém o loop principal que orquestra a simulação e os alertas.
- **Configurações como atributos**: Limites, intervalos e e-mail agora são atributos da classe, definidos na inicialização.

```python
import time
import random

class SistemaMonitoramento:
    def __init__(self, limite_critico, intervalo_check_segundos, email_alerta=None):
        """
        Inicializa o sistema de monitoramento com suas configurações.

        Args:
            limite_critico (float): O limite a partir do qual a condição é crítica.
            intervalo_check_segundos (int): Intervalo entre as verificações (em segundos).
            email_alerta (str, optional): Endereço de e-mail para o alerta simulado.
        """
        self.limite_critico = limite_critico
        self.intervalo_check_segundos = intervalo_check_segundos
        self.email_alerta = email_alerta
        print(f"[V2] Sistema Monitoramento inicializado. Limite crítico: {self.limite_critico}°C.")

    def _verificar_condicao_critica(self, valor_atual):
        """
        Verifica se o valor atual excede o limite crítico do sistema.

        Args:
            valor_atual (float): O valor a ser verificado (ex: temperatura).

        Returns:
            bool: True se a condição for crítica, False caso contrário.
        """
        print(f"[V2 Monitor] Valor atual: {valor_atual:.2f} (Limite crítico: {self.limite_critico:.2f})")
        return valor_atual > self.limite_critico

    def _disparar_alerta(self, mensagem_alerta, tipo_alerta='console'):
        """
        Dispara um alerta, seja no console ou simulando um e-mail.

        Args:
            mensagem_alerta (str): A mensagem a ser incluída no alerta.
            tipo_alerta (str): 'console' para imprimir no console, 'email_simulado' para simular e-mail.
        """
        if tipo_alerta == 'console':
            print(f"\n{'='*30}\n[V2 ALERTA CRÍTICO]: {mensagem_alerta}\n{'='*30}\n")
        elif tipo_alerta == 'email_simulado':
            if self.email_alerta:
                print(f"\n{'='*30}\n[V2 SIMULANDO ENVIO DE E-MAIL PARA {self.email_alerta}]:\nAssunto: ALERTA DE SISTEMA CRÍTICO (V2)\nCorpo: {mensagem_alerta}\n{'='*30}\n")
            else:
                print("[V2] Erro: E-mail de destino não configurado para alerta simulado.")
        else:
            print(f"[V2] Tipo de alerta '{tipo_alerta}' não reconhecido. Mensagem: {mensagem_alerta}")

    def iniciar_monitoramento(self):
        """
        Inicia o loop principal de monitoramento.
        """
        print(f"[V2] Iniciando monitoramento autônomo. Verificação a cada {self.intervalo_check_segundos} segundos. Pressione Ctrl+C para parar.")
        try:
            while True:
                # Simula a leitura de um sensor (temperatura)
                valor_do_sensor = random.uniform(20.0, 35.0) # Gera um valor aleatório

                if self._verificar_condicao_critica(valor_do_sensor):
                    mensagem = f"Superaquecimento detectado! Temperatura: {valor_do_sensor:.2f}°C, acima do limite de {self.limite_critico}°C."

                    self._disparar_alerta(mensagem, tipo_alerta='console')
                    self._disparar_alerta(mensagem, tipo_alerta='email_simulado')
                else:
                    print("[V2 Monitor] Condição normal.")

                time.sleep(self.intervalo_check_segundos)

        except KeyboardInterrupt:
            print("\n[V2] Monitoramento encerrado pelo usuário.")
        except Exception as e:
            print(f"[V2] Ocorreu um erro inesperado no loop principal: {e}")

# --- Exemplo de Uso da Versão Refatorada ---
if __name__ == '__main__':
    # Configurações para a nova versão
    NOVO_LIMITE_CRITICO = 32.0 # Exemplo: um limite diferente
    NOVO_INTERVALO_CHECK = 3  # Exemplo: um intervalo diferente
    NOVO_EMAIL_ALERTA = "monitoramento.v2@exemplo.com" # Seu e-mail aqui

    # Instancia o sistema de monitoramento
    monitor_v2 = SistemaMonitoramento(
        limite_critico=NOVO_LIMITE_CRITICO,
        intervalo_check_segundos=NOVO_INTERVALO_CHECK,
        email_alerta=NOVO_EMAIL_ALERTA
    )

    # Inicia o monitoramento
    monitor_v2.iniciar_monitoramento()
```
