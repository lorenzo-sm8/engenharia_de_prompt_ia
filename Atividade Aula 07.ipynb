```python
def calculadora():
    # Exibe uma mensagem de boas-vindas ao iniciar a calculadora
    print("Bem-vindo à Calculadora Simples!")

    # Loop infinito para manter a calculadora rodando até o usuário decidir sair
    while True:
        try:
            # Solicita e converte o primeiro número para float
            num1 = float(input("Digite o primeiro número: "))
            # Solicita e converte o segundo número para float
            num2 = float(input("Digite o segundo número: "))
            # Solicita a operação e remove espaços extras com .strip()
            operacao = input("Escolha a operação (+, -, *, /) ou 's' para sair: ").strip()

            # Verifica se o usuário deseja sair do programa
            if operacao == 's':
                print("Saindo da calculadora. Até mais!")
                # Sai do loop while True, encerrando a função
                break

            # Verifica a operação escolhida e realiza o cálculo correspondente
            if operacao == '+':
                resultado = num1 + num2
                print(f"O resultado da soma é: {resultado}")
            elif operacao == '-':
                resultado = num1 - num2
                print(f"O resultado da subtração é: {resultado}")
            elif operacao == '*':
                resultado = num1 * num2
                print(f"O resultado da multiplicação é: {resultado}")
            elif operacao == '/':
                # Verifica se o divisor não é zero para evitar erro de divisão por zero
                if num2 != 0:
                    resultado = num1 / num2
                    print(f"O resultado da divisão é: {resultado}")
                else:
                    # Mensagem de erro para divisão por zero
                    print("Erro: Divisão por zero não é permitida!")
            else:
                # Mensagem de erro para operação inválida
                print("Operação inválida. Por favor, escolha entre +, -, *, / ou 's'.")
        except ValueError:
            # Captura erro se a entrada do usuário não for um número válido
            print("Entrada inválida. Por favor, digite apenas números.")
        except Exception as e:
            # Captura qualquer outro erro inesperado que possa ocorrer
            print(f"Ocorreu um erro inesperado: {e}")

        # Adiciona uma linha em branco para melhor legibilidade na próxima interação
        print("\n")

# Chama a função da calculadora para iniciar o programa
calculadora()
```
